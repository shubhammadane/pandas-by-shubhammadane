**AI vs ML vs DL**
===================
What is AI? Give a real-time example.
------------------------------------------
AI is the technology that makes machines think and act like humans. It can make decisions, solve problems, and understand language.
Real-time: Google Maps traffic prediction, chatbots, voice assistants.

-----------------------------------------------

--------------------------------------

Real-time: Face recognition, medical image detection.

---------------------------------
AI is the broad field. ML is AI that learns from data. DL is ML using deep neural networks.

What is Linear Regression?
==================================

Linear Regression is a Machine Learning algorithm used to find a straight-line relationship between:

Input (X) ‚Üí independent variable

Output (Y) ‚Üí dependent variable

It answers questions like:

If X increases, how does Y change?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df =  pd.read_csv('/content/Dataset/placement.csv')
df.head()
```

```python
#scatter plot is a graph used to show the relationship between two numerical variables.
#X-axis ‚Üí CGPA
#Y-axis ‚Üí Package (Salary)
#Each dot = One student‚Äôs data

plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
# What is iloc ?
# ---------------
# iloc = Index Location
# It is used to select data from a DataFrame using row and column numbers (indexes) ‚Äî not names.
# Format:
# df.iloc[row_selection , column_selection]

```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='Black')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```

```python
m * 6.35 + b
```

```python
m * 100 + b
```
