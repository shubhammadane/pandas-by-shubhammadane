**AI vs ML vs DL**
===================
What is AI? Give a real-time example.
------------------------------------------
AI is the technology that makes machines think and act like humans. It can make decisions, solve problems, and understand language.
Real-time: Google Maps traffic prediction, chatbots, voice assistants.

What is ML? How is it used in real life?
-----------------------------------------------
Machine Learning is a part of AI where systems learn from data and improve automatically without programming rules.
Real-time: Gmail spam filter learns from emails and blocks spam.

What is Deep Learning?
--------------------------------------
Deep Learning is a subset of ML that uses multi-layer neural networks to learn complex patterns from large data.
Real-time: Face recognition, medical image detection.


Difference between AI, ML, DL?
---------------------------------
AI is the broad field. ML is AI that learns from data. DL is ML using deep neural networks.
Real-time:

AI ‚Üí Chatbot

ML ‚Üí Movie recommendation

DL ‚Üí Self-driving car vision
