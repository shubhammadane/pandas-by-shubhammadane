**Retrieving Multiple Rows**

**1. Using `iloc` for Positional Indexing:**

To retrieve rows by their integer position, you can use `iloc` with slicing. For example, to get the first three rows (rows with index 0, 1, and 2):

```python
print("\nFirst three rows using df.iloc[0:3]:")
display(df.iloc[0:3])
```

You can also specify a list of integer positions to retrieve specific non-contiguous rows:

```python
print("\nRows at index 1 and 4 using df.iloc[[1, 4]]:")
display(df.iloc[[1, 4]])
```

**2. Using `loc` for Label-based or Boolean Indexing:**

To retrieve rows based on conditions on column values, you can use `loc` with boolean indexing. For example, to get all employees with an 'Age' greater than 24:

```python
print("\nEmployees with Age > 24 using df.loc[df['Age'] > 24]:")
display(df.loc[df['Age'] > 24])
```

You can also combine multiple conditions using logical operators (`&` for AND, `|` for OR):

```python
print("\nEmployees named 'Shubham' or with Salary > 70000:")
display(df.loc[(df['Name'] == 'Shubham') | (df['Salary'] > 70000)])
```
