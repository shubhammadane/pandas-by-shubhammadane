```python
# Select the first row (index 0) using .iloc[]
print("\nFirst row using df.iloc[0]:")
display(df.iloc[0])
```

```python
# Select the last row using .iloc[-1]
print("\nLast row using df.iloc[-1]:")
display(df.iloc[-1])
```

```python
# Alternatively, you can use .head(1) for the first row
print("\nFirst row using df.head(1):")
display(df.head(1))
```

```python
# And .tail(1) for the last row
print("\nLast row using df.tail(1):")
display(df.tail(1))
```
