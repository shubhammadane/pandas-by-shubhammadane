```python
import matplotlib.pyplot as plt
#matplotlib --->Main Package
#.pyplotb--->module(provide the pre function or code)
#as --->Alias Name

students = ['A', 'B', 'C', 'D']
marks = [80, 65, 90, 75]

plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks Visualization")

plt.show()
```
