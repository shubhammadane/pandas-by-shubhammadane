```python
#Check Positive Number
num = float(input("Enter a number: "))
if num > 0:
   print("Positive number")
elif num == 0:
   print("Zero")
else:
   print("Negative number")
```

```python
#Check Even and Odd Number
num = int(input("Enter a number: "))
if (num % 2) == 0:
   print("{0} is Even".format(num))
else:
   print("{0} is Odd".format(num))
```

```python
#Check Greater Number
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print(f"The largest number is {largest}")
```

```python
#Largest of Two Numbers
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if a > b:
    print(f"{a} is largest")
else:
    print(f"{b} is largest")
```

```python
#Check Age for Voting
age = int(input("Enter your age: "))
if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
#Grade System
marks = int(input("Enter your marks: "))
if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
```

```python
#Check Letter is Vowel
letter = input("Enter a letter: ")
if letter.lower() in ('a', 'e', 'i', 'o', 'u'):
    print(f"{letter} is a vowel")
else:
    print(f"{letter} is a consonant")
```

```python
#Check Character is Digit
char = input("Enter a character: ")
if char.isdigit():
    print(f"{char} is a digit")
else:
    print(f"{char} is not a digit")
```

```python
#Check Leap Year
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
#Switch Case – Print Day Name
day = int(input("Enter a day number: "))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case _:
        print("Invalid Day")
```

```python
#Switch Case – Simple Calculator
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter an operator (+, -, *, /): ")

match op:
    case "+":
        print(a + b)
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/":
        # Added a check for division by zero
        if b != 0:
            print(a / b)
        else:
            print("Error: Division by zero")
    case _:
        print("Invalid Operation")
```

```python
#Switch Case – Grade System
grade = input("Enter your grade: ")

match grade:
    case  "A":
        print("Excellent")
    case "B":
        print("Good")
    case "C":
        print("Average")
    case "D":
        print("Fail")
    case _:
        print("Invalid Grade")
```
