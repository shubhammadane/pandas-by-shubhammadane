What is pixel?
===============
A pixel is the smallest unit of a digital image that stores color information. In color images, each pixel is represented using RGB values ranging from 0 to 255. An image is stored as a matrix, and a color image has the shape (Height, Width, Channels)

An image is made up of thousands or millions of pixels

Each pixel stores color information

When many pixels come together → we see a complete image

 Real-time example:
 -------------------
Think of a mosaic photo made from tiny colored tiles.


Each tile = one pixel


All tiles together = image

RGB Values (0–255)
=====================

Each pixel’s color is represented using RGB:

R → Red

G → Green

B → Blue

Each value ranges from 0 to 255

Color	RGB Value
----------------

Black	(0, 0, 0)


White	(255, 255, 255)


Red	(255, 0, 0)


Green	(0, 255, 0)


Blue	(0, 0, 255)


Yellow	(255, 255, 0)


Real-time example:
--------------------------
On your mobile camera, when you capture a photo, the camera stores RGB values for every pixel to recreate the image.

Image as a Matrix
============================
In computer vision, an image is stored as a matrix (array of numbers).

Grayscale image → 2D matrix

Color image → 3D matrix

Example (Grayscale):

[
 [  0,  50, 120],
 [ 30, 200, 255],
 [ 10,  80, 160]
]

Each number = pixel intensity

Shape → (H, W, C)
------------------
For a color image, the shape is written as:

(H, W, C)


Where:

H → Height (number of rows)

W → Width (number of columns)

C → Channels (3 for RGB)


Example:
--------------
Image shape = (256, 256, 3)


 Image height = 256 pixels
 Image width = 256 pixels
 3 color channels (R, G, B)

Real-time example:
---------------------------
A Full HD image:

(1080, 1920, 3)


Real-Time Applications
==========================
✔ Face recognition (each face = pixel values)
✔ Medical imaging (X-ray pixels show intensity)
✔ Self-driving cars (road images analyzed pixel-wise)
✔ CCTV cameras (motion detection using pixel changes
