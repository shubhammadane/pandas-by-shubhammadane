What is Pandas?
==============
Pandas is a Python library used for data manipulation and analysis. It provides easy-to-use data structures and functions to work with structured data, like tables (similar to Excel spreadsheets or SQL tables).

Key features of Pandas:
=======================

Works with tabular data (rows and columns) using DataFrames.

Handles missing data gracefully.

Offers fast data filtering, grouping, merging, and aggregation.

Makes it easy to read/write data from CSV, Excel, SQL, JSON, etc.

Why use Pandas?
=================
Simplifies data cleaning and analysis.

Fast and efficient for large datasets.

Integrates well with NumPy, Matplotlib, and SciPy for scientific computing.

What is a DataFrame?
=====================
A DataFrame is a 2-dimensional data structure in Pandas used to store data in rows and columns, just like:

Excel sheet

Database table

CSV file

Each column can have a different data type (int, float, string, etc.).

Why DataFrame is used?
========================
To store structured data

Easy data analysis & manipulation

Supports filtering, sorting, grouping, and aggregation

Handles missing values

```python
#Real Time Example
import pandas as pd
data = {
    'StudentName': ['Shubham', 'Raj', 'Suyash', 'Vaibhav', 'Kiran', 'Neha'],
    'mark': [85, 90, 78, 88, 92, 80]
}
#Crate DataFrame
df = pd.DataFrame(data)
print(df)
#Read 5 rows
print("-------------------")
print(df.head(2))
#Read last 5 rows
print("-------------------")
print(df.tail(-2))
print("--------------------")
print(df.shape)
print("--------------------")
print(df.columns)
print('-------------------')
ren = df.rename(columns={'mark':'stdMark'},inplace=True)
print(ren)
```

```python
df.info()
# What it does
# Gives structure of the DataFrame
# Shows data types
# Shows non-null values
# Shows memory usage
```

```python
df.describe()

# What it does
# Gives statistical summary
# Works on numerical columns by default
# Statistics Included
# count
# mean
# std (standard deviation)
# min
# 25%, 50%, 75% (quartiles)
# max
```
